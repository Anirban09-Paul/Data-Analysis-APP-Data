# Now we create a new data frame there only numeric data types

#   Data Visualization 

we use Heatmap to find the correlation between each and every features 

```python
plt.figure(figsize=(12,6))
sns.heatmap(app_data2.corr(),annot=True,cmap='coolwarm')
plt.title('Correlation matrix of app_data2',fontsize=20)
```

```python
sns.pairplot(app_data2,hue='enrolled')
```

 It will count how many people used premium features, 'Orange' means Subscribe and 'Blue' means Unsubscribe

```python
sns.countplot(x='used_premium_feature',data=app_data2)
print('Not subscribed features= ',(app_data2.used_premium_feature<1).sum(),'Out of 50000')
print('Subscribed features= ',50000-(app_data2.used_premium_feature<1).sum(),'Out of 50000')
```
